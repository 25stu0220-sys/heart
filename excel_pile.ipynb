```python
import matplotlib.pyplot as plt
import numpy as np

# 학생 데이터 예시 (입력 대신 고정값 사용)
students = [
    {'이름': 'son', '국어': 99, '수학': 88, '과학': 77},
    {'이름': 'kane', '국어': 80, '수학': 90, '과학': 71},
    {'이름': 'ali', '국어': 67, '수학': 78, '과학': 89},
    {'이름': 'eriksen', '국어': 45, '수학': 56, '과학': 77},
    {'이름': 'dier', '국어': 34, '수학': 54, '과학': 55},
]

num_students = len(students)

# 학생별 합계, 평균 계산 추가
for s in students:
    total = s['국어'] + s['수학'] + s['과학']
    avg = total / 3
    s['합계'] = total
    s['평균'] = avg

# 과목별 평균 계산
avg_korean = np.mean([s['국어'] for s in students])
avg_math = np.mean([s['수학'] for s in students])
avg_science = np.mean([s['과학'] for s in students])

# 테이블 데이터 준비
columns = ('name', 'language', 'math', 'science', 'total', 'mean')
rows = [(
    s['이름'],
    s['국어'],
    s['수학'],
    s['과학'],
    s['합계'],
    f"{s['평균']:.2f}"
) for s in students]

# 평균 행 추가
rows.append((
    'average',
    f"{avg_korean:.2f}",
    f"{avg_math:.2f}",
    f"{avg_science:.2f}",
    '',
    ''
))

# 시각화
fig, ax = plt.subplots(figsize=(10, 3))
ax.axis('off')  # 축 제거

# 표 그리기
table = ax.table(cellText=rows, colLabels=columns, loc='center', cellLoc='center')

table.auto_set_font_size(False)
table.set_fontsize(12)
table.scale(1, 2)  # 표 크기 조절

# 헤더 스타일 지정
for (row, col), cell in table.get_celld().items():
    if row == 0:
        cell.set_fontsize(14)
        cell.set_text_props(weight='bold', color='white')
        cell.set_facecolor('#40466e')
    elif row == len(rows):
        cell.set_facecolor('#e1e1e1')

plt.title('student score', fontsize=16, pad=20)
plt.show()
```
